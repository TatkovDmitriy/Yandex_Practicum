#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию.

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные.

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений.

Чтобы разобраться, команда решила провести A/B-эксперимент.

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года.
Пользователей случайным образом разделили на две равные группы:
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках.

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита.

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик:
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.1.** Анализ новых пользователей.
- Изучите динамику привлечения новых пользователей в приложение.

````python
# 📊 Подсчёт количества новых пользователей по дате регистрации
new_users_daily = (
    hist_data[['user_id', 'first_dt']]
    .drop_duplicates()
    .groupby('first_dt')
    .agg(users_count=('user_id', 'nunique'))
    .reset_index()
)

# 📈 Построение графика
plt.figure(figsize=(12, 5))
sns.lineplot(data=new_users_daily, x='first_dt', y='users_count', marker='o')
plt.title(' Динамика привлечения новых пользователей')
plt.xlabel('Дата регистрации')
plt.ylabel('Количество новых пользователей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📋 Таблица с данными
new_users_daily.tail(10)  # можно заменить на display(new_users_daily) для полного вывода
````

````python
# 📊 Подсчёт новых пользователей по дате и платформе
new_users_by_platform = (
    hist_data[['user_id', 'first_dt', 'platform']]
    .drop_duplicates()
    .groupby(['first_dt', 'platform'])
    .agg(users_count=('user_id', 'nunique'))
    .reset_index()
)

# 📉 Добавление сглаженного значения (rolling 7)
new_users_by_platform['users_count_smooth'] = (
    new_users_by_platform
    .groupby('platform')['users_count']
    .transform(lambda x: x.rolling(7, min_periods=1).mean())
)

# 📈 График: динамика по платформам с 7-дневным сглаживанием
plt.figure(figsize=(12, 5))
sns.lineplot(
    data=new_users_by_platform,
    x='first_dt',
    y='users_count_smooth',
    hue='platform'
)
plt.title('Динамика привлечения новых пользователей по платформам (скользящее среднее 7 дней)')
plt.xlabel('Дата регистрации')
plt.ylabel('Количество пользователей (7-дневное сглаживание)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📋 Таблица со сглаженными данными
new_users_by_platform.tail(10)  # или display(new_users_by_platform)
````

````python
# 📊 Подсчёт новых пользователей по дате и стране
new_users_by_country = (
    hist_data[['user_id', 'first_dt', 'country_code']]
    .drop_duplicates()
    .groupby(['first_dt', 'country_code'])
    .agg(users_count=('user_id', 'nunique'))
    .reset_index()
)

# 📉 Добавление сглаженных значений (rolling 7)
new_users_by_country['users_count_smooth'] = (
    new_users_by_country
    .groupby('country_code')['users_count']
    .transform(lambda x: x.rolling(7, min_periods=1).mean())
)

# 📈 График: динамика по странам с 7-дневным сглаживанием
plt.figure(figsize=(12, 5))
sns.lineplot(
    data=new_users_by_country,
    x='first_dt',
    y='users_count_smooth',
    hue='country_code'
)
plt.title('Динамика привлечения новых пользователей по странам (скользящее среднее 7 дней)')
plt.xlabel('Дата регистрации')
plt.ylabel('Количество пользователей (7-дневное сглаживание)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📋 Таблица со сглаженными значениями
new_users_by_country.tail(10)  # или display(new_users_by_country)
````

````python
# 🔍 Определим последовательность всех ключевых шагов в пользовательском пути
funnel_steps = [
    'install / open_web',
    'introduction',
    'main_page',
    'onboarding_complete',
    'registration',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# Убедимся, что события отсортированы по логике пути
hist_data['event_name'] = pd.Categorical(hist_data['event_name'], categories=funnel_steps, ordered=True)

# 📊 Расчёт количества уникальных пользователей на каждом этапе
funnel_all = (
    hist_data[hist_data['event_name'].isin(funnel_steps)]
    .drop_duplicates(subset=['user_id', 'event_name'])  # чтобы не учитывать повторные действия
    .groupby('event_name')['user_id']
    .nunique()
    .reset_index(name='users_count')
)

# 🎯 Добавим конверсии: от первого шага и пошаговую
first_step_count = funnel_all['users_count'].iloc[0]
funnel_all['classic_cr'] = funnel_all['users_count'] / first_step_count
funnel_all['step_cr'] = funnel_all['users_count'] / funnel_all['users_count'].shift(1)

# 💡 Форматируем как проценты
funnel_all['classic_cr_pct'] = (funnel_all['classic_cr'] * 100).round(2).astype(str) + '%'
funnel_all['step_cr_pct'] = (funnel_all['step_cr'] * 100).round(2).astype(str) + '%'

# 📈 Визуализация классической воронки
plt.figure(figsize=(10, 5))
sns.barplot(data=funnel_all, x='event_name', y='classic_cr')
plt.title('Классическая воронка (от первого шага)')
plt.ylabel('Конверсия от первого шага')
plt.xlabel('Событие')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📈 Визуализация пошаговой воронки
plt.figure(figsize=(10, 5))
sns.barplot(data=funnel_all, x='event_name', y='step_cr')
plt.title('Пошаговая воронка (от предыдущего шага)')
plt.ylabel('Конверсия от предыдущего шага')
plt.xlabel('Событие')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📋 Таблица для отчёта
funnel_all[['event_name', 'users_count', 'classic_cr_pct', 'step_cr_pct']]
````

- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.  

````python
# 🔄 Только активы с валидным уровнем риска (исключаем unknown)
assets_df = hist_data[
    hist_data['risk_level'].isin(['low', 'medium', 'high'])
].copy()

# 🧱 Берём первый актив с уровнем риска
first_assets = (
    assets_df.sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()
    [['user_id', 'risk_level']]
)

# 📌 Отметим, кто сделал второй депозит
second_deposit_users = (
    hist_data[hist_data['event_name'] == 'second_deposit']['user_id'].unique()
)

first_assets['made_second_deposit'] = first_assets['user_id'].isin(second_deposit_users)

#  Считаем конверсии по уровням риска
risk_conv = (
    first_assets
    .groupby('risk_level')
    .agg(users_total=('user_id', 'count'),
         converted=('made_second_deposit', 'sum'))
    .reset_index()
)

risk_conv['conversion_rate'] = risk_conv['converted'] / risk_conv['users_total']
risk_conv = risk_conv.sort_values(by='conversion_rate', ascending=False)

# 📈 Визуализация (опционально)
plt.figure(figsize=(8, 4))
sns.barplot(data=risk_conv, x='risk_level', y='conversion_rate')
plt.title(' Конверсия во второй депозит по уровню риска первого актива')
plt.ylabel('Конверсия')
plt.xlabel('Уровень риска')
plt.tight_layout()
plt.show()

# 📋 Таблица
risk_conv
````

````python
# 📌 Полный путь пользователя
funnel_steps = [
    'install / open_web',
    'introduction',
    'main_page',
    'onboarding_complete',
    'registration',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# Отфильтруем события по этим шагам
abt_funnel = abt_data[abt_data['event_name'].isin(funnel_steps)].copy()

# Установим порядок
abt_funnel['event_name'] = pd.Categorical(abt_funnel['event_name'], categories=funnel_steps, ordered=True)

# Удалим дубликаты
abt_funnel_unique = abt_funnel.drop_duplicates(subset=['user_id', 'event_name', 'group'])

# 📊 Посчитаем количество пользователей на каждом шаге по группам
funnel_by_group = (
    abt_funnel_unique
    .groupby(['group', 'event_name'])['user_id']
    .nunique()
    .reset_index(name='users_count')
)

# 📈 Пивот-таблица для расчёта конверсий
funnel_pivot = funnel_by_group.pivot(index='event_name', columns='group', values='users_count').reset_index()

# 📉 Расчёт пошаговой и классической конверсии для каждой группы
for grp in ['control', 'test']:
    funnel_pivot[f'{grp}_step_cr'] = funnel_pivot[grp] / funnel_pivot[grp].shift(1)
    funnel_pivot[f'{grp}_total_cr'] = funnel_pivot[grp] / funnel_pivot[grp].iloc[0]

# 📊 Визуализация: абсолютные значения
plt.figure(figsize=(12, 5))
sns.barplot(data=funnel_by_group, x='event_name', y='users_count', hue='group')
plt.title('Распределение пользователей по шагам воронки (по группам)')
plt.xlabel('Событие')
plt.ylabel('Уникальные пользователи')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📊 Визуализация пошаговой конверсии
cr_plot = funnel_pivot.melt(
    id_vars='event_name',
    value_vars=['control_step_cr', 'test_step_cr'],
    var_name='group_step',
    value_name='conversion_rate'
)
cr_plot['group'] = cr_plot['group_step'].str.extract(r'(control|test)')

plt.figure(figsize=(12, 5))
sns.barplot(data=cr_plot, x='event_name', y='conversion_rate', hue='group')
plt.title('Пошаговая конверсия (step conversion rate) по группам')
plt.xlabel('Шаг воронки')
plt.ylabel('Конверсия')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 📋 Таблица с итогами
funnel_pivot
````

````python
# Только валидные уровни риска
valid_assets = abt_data[
    abt_data['risk_level'].isin(['low', 'medium', 'high'])
].copy()

# Первый актив каждого пользователя
first_assets = (
    valid_assets.sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()
    [['user_id', 'group', 'risk_level']]
)
# Массив user_id с событием second_deposit
second_deposit_users = abt_data[
    abt_data['event_name'] == 'second_deposit'
]['user_id'].unique()

first_assets['second_deposit'] = first_assets['user_id'].isin(second_deposit_users)
# Агрегируем
risk_conv_grouped = (
    first_assets
    .groupby(['group', 'risk_level'])
    .agg(users_total=('user_id', 'count'),
         converted=('second_deposit', 'sum'))
    .reset_index()
)

# Конверсия
risk_conv_grouped['conversion_rate'] = risk_conv_grouped['converted'] / risk_conv_grouped['users_total']
plt.figure(figsize=(10, 5))
sns.barplot(data=risk_conv_grouped, x='risk_level', y='conversion_rate', hue='group')
plt.title('Конверсия во второй депозит по уровню риска и группе')
plt.ylabel('Конверсия')
plt.xlabel('Уровень риска первого актива')
plt.tight_layout()
plt.show()
risk_conv_grouped
````

````python
# Только пользователи, совершившие хотя бы один депозит
paid_users = deposit_all_users[deposit_all_users['amount'] > 0].copy()

# Разделим по группам
control_group = paid_users[paid_users['group'] == 'control']['amount'].values
test_group = paid_users[paid_users['group'] == 'test']['amount'].values
import numpy as np

# Фиксируем сид
np.random.seed(42)

# Настройки
n_iterations = 1000
percentiles = [25, 50, 75]
bootstrap_results = {}

for p in percentiles:
    boot_diffs = []
    for _ in range(n_iterations):
        control_sample = np.random.choice(control_group, size=len(control_group), replace=True)
        test_sample = np.random.choice(test_group, size=len(test_group), replace=True)

        diff = np.percentile(test_sample, p) - np.percentile(control_sample, p)
        boot_diffs.append(diff)

    boot_diffs = np.array(boot_diffs)
    ci = np.percentile(boot_diffs, [2.5, 97.5])

    bootstrap_results[p] = {
        'control_percentile': np.percentile(control_group, p),
        'test_percentile': np.percentile(test_group, p),
        'mean_diff': np.mean(boot_diffs),
        'ci_low': ci[0],
        'ci_high': ci[1],
        'distribution': boot_diffs
    }
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))

for i, p in enumerate(percentiles):
    plt.subplot(1, 3, i + 1)
    plt.hist(bootstrap_results[p]['distribution'], bins=50, alpha=0.7)
    plt.axvline(bootstrap_results[p]['ci_low'], color='red', linestyle='--')
    plt.axvline(bootstrap_results[p]['ci_high'], color='red', linestyle='--')
    plt.axvline(0, color='black')
    plt.title(f'Разница {p}-го перцентиля\n95% CI: [{bootstrap_results[p]["ci_low"]:.2f}, {bootstrap_results[p]["ci_high"]:.2f}]')
    plt.xlabel('Разница (test - control)')
    plt.ylabel('Частота')

plt.tight_layout()
plt.show()
import pandas as pd

summary = pd.DataFrame([
    {
        'percentile': p,
        'control': bootstrap_results[p]['control_percentile'],
        'test': bootstrap_results[p]['test_percentile'],
        'mean_diff': bootstrap_results[p]['mean_diff'],
        'ci_low': bootstrap_results[p]['ci_low'],
        'ci_high': bootstrap_results[p]['ci_high']
    }
    for p in percentiles
])

summary
````
