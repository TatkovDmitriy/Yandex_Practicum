````python
# üì¶ –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üß™ –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
from scipy import stats
from scipy.stats import ttest_ind, mannwhitneyu, bootstrap

# üìä –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤
sns.set_theme(style='whitegrid', palette='pastel')
%matplotlib inline

# üóÇÔ∏è –†–∞–±–æ—Ç–∞ —Å –¥–∞—Ç–∞–º–∏
from datetime import datetime, timedelta

# üìà –†–µ–≥—Ä–µ—Å—Å–∏—è –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ (–µ—Å–ª–∏ –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# üßº –ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏—è
import warnings
warnings.filterwarnings('ignore')
````

````python
# üì• –®–∞–≥ 1. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
# (–≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –æ–¥–∏–Ω —Ä–∞–∑, –æ—Ç–¥–µ–ª—å–Ω–æ –æ—Ç –∞–Ω–∞–ª–∏–∑–∞)
hist_data = pd.read_csv('https://code.s3.yandex.net/datasets/pa_sollmafin_hist.csv')
````

````python
# üîç –®–∞–≥ 2. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤

# –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ç–∏–ø–∞—Ö –∏ —Ä–∞–∑–º–µ—Ä–µ
hist_data.info()

# –ü—Ä–∏–º–µ—Ä—ã —Å—Ç—Ä–æ–∫
display(hist_data.head())

# –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ (–≤ —Ç.—á. –¥–ª—è object)
display(hist_data.describe(include='all'))

# –ü—Ä–æ–ø—É—Å–∫–∏ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º
display(hist_data.isna().sum())
````

````python
# üßπ –®–∞–≥ 1. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ñ–æ—Ä–º–∞—Ç–æ–≤ –∏ —É–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

# –î–∞—Ç–∞-–≤—Ä–µ–º—è ‚Üí datetime
hist_data['first_ts'] = pd.to_datetime(hist_data['first_ts'])
hist_data['first_dt'] = pd.to_datetime(hist_data['first_dt'])
hist_data['event_ts'] = pd.to_datetime(hist_data['event_ts'])

# –£–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
hist_data = hist_data.drop_duplicates()
````

````python
# üîç –û–ø—Ä–µ–¥–µ–ª–∏–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –∫–ª—é—á–µ–≤—ã—Ö —à–∞–≥–æ–≤ –≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º –ø—É—Ç–∏
funnel_steps = [
    'install / open_web',
    'introduction',
    'main_page',
    'onboarding_complete',
    'registration',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ —Å–æ–±—ã—Ç–∏—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã –ø–æ –ª–æ–≥–∏–∫–µ –ø—É—Ç–∏
hist_data['event_name'] = pd.Categorical(hist_data['event_name'], categories=funnel_steps, ordered=True)

# üìä –†–∞—Å—á—ë—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ
funnel_all = (
    hist_data[hist_data['event_name'].isin(funnel_steps)]
    .drop_duplicates(subset=['user_id', 'event_name'])  # —á—Ç–æ–±—ã –Ω–µ —É—á–∏—Ç—ã–≤–∞—Ç—å –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
    .groupby('event_name')['user_id']
    .nunique()
    .reset_index(name='users_count')
)

# üéØ –î–æ–±–∞–≤–∏–º –∫–æ–Ω–≤–µ—Ä—Å–∏–∏: –æ—Ç –ø–µ—Ä–≤–æ–≥–æ —à–∞–≥–∞ –∏ –ø–æ—à–∞–≥–æ–≤—É—é
first_step_count = funnel_all['users_count'].iloc[0]
funnel_all['classic_cr'] = funnel_all['users_count'] / first_step_count
funnel_all['step_cr'] = funnel_all['users_count'] / funnel_all['users_count'].shift(1)

# üí° –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –∫–∞–∫ –ø—Ä–æ—Ü–µ–Ω—Ç—ã
funnel_all['classic_cr_pct'] = (funnel_all['classic_cr'] * 100).round(2).astype(str) + '%'
funnel_all['step_cr_pct'] = (funnel_all['step_cr'] * 100).round(2).astype(str) + '%'

# üìà –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π –≤–æ—Ä–æ–Ω–∫–∏
plt.figure(figsize=(10, 5))
sns.barplot(data=funnel_all, x='event_name', y='classic_cr')
plt.title('–ö–ª–∞—Å—Å–∏—á–µ—Å–∫–∞—è –≤–æ—Ä–æ–Ω–∫–∞ (–æ—Ç –ø–µ—Ä–≤–æ–≥–æ —à–∞–≥–∞)')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è –æ—Ç –ø–µ—Ä–≤–æ–≥–æ —à–∞–≥–∞')
plt.xlabel('–°–æ–±—ã—Ç–∏–µ')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üìà –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ—à–∞–≥–æ–≤–æ–π –≤–æ—Ä–æ–Ω–∫–∏
plt.figure(figsize=(10, 5))
sns.barplot(data=funnel_all, x='event_name', y='step_cr')
plt.title('–ü–æ—à–∞–≥–æ–≤–∞—è –≤–æ—Ä–æ–Ω–∫–∞ (–æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —à–∞–≥–∞)')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —à–∞–≥–∞')
plt.xlabel('–°–æ–±—ã—Ç–∏–µ')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üìã –¢–∞–±–ª–∏—Ü–∞ –¥–ª—è –æ—Ç—á—ë—Ç–∞
funnel_all[['event_name', 'users_count', 'classic_cr_pct', 'step_cr_pct']]
````

## üîç –£–∑–∫–∏–µ –º–µ—Å—Ç–∞:

* –ü–∞–¥–µ–Ω–∏–µ —Å `registration` –¥–æ `first_deposit` (**‚àí25 –ø.–ø.**) ‚Äî **–∫–ª—é—á–µ–≤–∞—è —Ç–æ—á–∫–∞ –æ—Ç—Ç–æ–∫–∞**: –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø—Ä–æ—à–ª–∏ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—é, –Ω–æ –Ω–µ —Å–æ–≤–µ—Ä—à–∏–ª–∏ –ø–µ—Ä–≤–æ–µ –≤–ª–æ–∂–µ–Ω–∏–µ.
* –¢–∞–∫–∂–µ –∑–Ω–∞—á–∏–º –ø—Ä–æ–≤–∞–ª –Ω–∞ —ç—Ç–∞–ø–µ `asset_purchase`: –±–æ–ª–µ–µ **40% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –¥–µ–ø–æ–∑–∏—Ç–æ–º –Ω–µ —Å–æ–≤–µ—Ä—à–∞—é—Ç –ø–æ–∫—É–ø–∫—É –∞–∫—Ç–∏–≤–∞**, —á—Ç–æ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ **–Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω—É—é –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ—Å—Ç—å –∏–ª–∏ –Ω–µ–ø–æ–Ω–∏–º–∞–Ω–∏–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ —à–∞–≥–∞**.

## üß© –í—ã–≤–æ–¥:

–û–Ω–±–æ—Ä–¥–∏–Ω–≥ –∏ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç–∞—é—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ, –Ω–æ **–æ—Å–Ω–æ–≤–Ω—ã–µ –ø–æ—Ç–µ—Ä–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥—è—Ç –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ ‚Äî –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ –∫ —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã–º –¥–µ–π—Å—Ç–≤–∏—è–º**. –≠—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ:

* —á–∞—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π **–Ω–µ –≥–æ—Ç–æ–≤–∞ –∏–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å —Å—Ä–∞–∑—É –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏**;
* –ª–∏–±–æ **–Ω–µ –ø–æ–Ω–∏–º–∞–µ—Ç –ø—Ä–æ–¥—É–∫—Ç–æ–≤—É—é –º–µ—Ö–∞–Ω–∏–∫—É –ø–æ–∫—É–ø–∫–∏ –∞–∫—Ç–∏–≤–æ–≤**;
* –ª–∏–±–æ **—Å–æ–º–Ω–µ–≤–∞–µ—Ç—Å—è –≤ –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç–∏**.

üìå –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –ø—Ä–æ–≤–µ—Å—Ç–∏ **–∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤—å—é** –∏–ª–∏ A/B-—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —É–ø—Ä–æ—â—ë–Ω–Ω–æ–≥–æ —Å—Ü–µ–Ω–∞—Ä–∏—è –º–µ–∂–¥—É —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–µ–π –∏ –ø–µ—Ä–≤—ã–º –¥–µ–ø–æ–∑–∏—Ç–æ–º.


**–ó–∞–¥–∞—á–∞ 2.3.** –í–ª–∏—è–Ω–∏–µ —É—Ä–æ–≤–Ω—è —Ä–∏—Å–∫–∞ –∞–∫—Ç–∏–≤–∞ –Ω–∞ –æ—Ç–∫—Ä—ã—Ç–∏–µ –≤—Ç–æ—Ä–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞.
- –ò–∑—É—á–∏—Ç–µ, –∫–∞–∫ —É—Ä–æ–≤–µ–Ω—å —Ä–∏—Å–∫–∞ –ø–µ—Ä–≤–æ–≥–æ –∞–∫—Ç–∏–≤–∞ –≤–ª–∏—è–µ—Ç –Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏—é –≤–æ –≤—Ç–æ—Ä–æ–π –¥–µ–ø–æ–∑–∏—Ç.  

````python
# üîÑ –¢–æ–ª—å–∫–æ –∞–∫—Ç–∏–≤—ã —Å –≤–∞–ª–∏–¥–Ω—ã–º —É—Ä–æ–≤–Ω–µ–º —Ä–∏—Å–∫–∞ (–∏—Å–∫–ª—é—á–∞–µ–º unknown)
assets_df = hist_data[
    hist_data['risk_level'].isin(['low', 'medium', 'high'])
].copy()

# üß± –ë–µ—Ä—ë–º –ø–µ—Ä–≤—ã–π –∞–∫—Ç–∏–≤ —Å —É—Ä–æ–≤–Ω–µ–º —Ä–∏—Å–∫–∞
first_assets = (
    assets_df.sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()
    [['user_id', 'risk_level']]
)

# üìå –û—Ç–º–µ—Ç–∏–º, –∫—Ç–æ —Å–¥–µ–ª–∞–ª –≤—Ç–æ—Ä–æ–π –¥–µ–ø–æ–∑–∏—Ç
second_deposit_users = (
    hist_data[hist_data['event_name'] == 'second_deposit']['user_id'].unique()
)

first_assets['made_second_deposit'] = first_assets['user_id'].isin(second_deposit_users)

#  –°—á–∏—Ç–∞–µ–º –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –ø–æ —É—Ä–æ–≤–Ω—è–º —Ä–∏—Å–∫–∞
risk_conv = (
    first_assets
    .groupby('risk_level')
    .agg(users_total=('user_id', 'count'),
         converted=('made_second_deposit', 'sum'))
    .reset_index()
)

risk_conv['conversion_rate'] = risk_conv['converted'] / risk_conv['users_total']
risk_conv = risk_conv.sort_values(by='conversion_rate', ascending=False)

# üìà –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)
plt.figure(figsize=(8, 4))
sns.barplot(data=risk_conv, x='risk_level', y='conversion_rate')
plt.title(' –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤–æ –≤—Ç–æ—Ä–æ–π –¥–µ–ø–æ–∑–∏—Ç –ø–æ —É—Ä–æ–≤–Ω—é —Ä–∏—Å–∫–∞ –ø–µ—Ä–≤–æ–≥–æ –∞–∫—Ç–∏–≤–∞')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å —Ä–∏—Å–∫–∞')
plt.tight_layout()
plt.show()

# üìã –¢–∞–±–ª–∏—Ü–∞
risk_conv
````

* –£—Ä–æ–≤–µ–Ω—å –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ —Å–Ω–∏–∂–∞–µ—Ç—Å—è –ø–æ –º–µ—Ä–µ —Ä–æ—Å—Ç–∞ —Ä–∏—Å–∫–∞:
  - **`medium`** ‚Äî 42,9%,
  - **`high`** ‚Äî –ª–∏—à—å **26,4%**.
* –≠—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –≥–∏–ø–æ—Ç–µ–∑—É: **–∏–Ω–≤–µ—Å—Ç–æ—Ä—ã, –Ω–∞—á–∏–Ω–∞—é—â–∏–µ —Å —Ä–∏—Å–∫–æ–≤–∞–Ω–Ω—ã—Ö –∞–∫—Ç–∏–≤–æ–≤, —Ç–µ—Ä—è—é—Ç –º–æ—Ç–∏–≤–∞—Ü–∏—é –ø—Ä–æ–¥–æ–ª–∂–∞—Ç—å**, –≤–µ—Ä–æ—è—Ç–Ω–æ, –∏–∑-–∑–∞ —É–±—ã—Ç–∫–æ–≤ –∏–ª–∏ —Å—Ç—Ä–µ—Å—Å–∞.
* –ù–∞–ø—Ä–æ—Ç–∏–≤, **–±–æ–ª–µ–µ –æ—Å—Ç–æ—Ä–æ–∂–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏** —á—É–≤—Å—Ç–≤—É—é—Ç —Å–µ–±—è —É–≤–µ—Ä–µ–Ω–Ω–æ –∏ —Å –±–æ–ª—å—à–µ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç—Å—è –¥–ª—è –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–π.

üìå –í—ã–≤–æ–¥: –æ–Ω–±–æ—Ä–¥–∏–Ω–≥ —Å –∞–∫—Ü–µ–Ω—Ç–æ–º –Ω–∞ –æ—Å–æ–∑–Ω–∞–Ω–Ω–æ—Å—Ç—å –∏ —Ä–∏—Å–∫–∏ –æ—Å–æ–±–µ–Ω–Ω–æ –≤–∞–∂–µ–Ω –¥–ª—è –Ω–æ–≤–∏—á–∫–æ–≤, —Å–∫–ª–æ–Ω–Ω—ã—Ö –≤—ã–±–∏—Ä–∞—Ç—å `high`-—Ä–∏—Å–∫. –û–Ω –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å **—É–¥–µ—Ä–∂–∞—Ç—å —ç—Ç–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –ø–æ–≤—ã—Å–∏—Ç—å –∏—Ö –∫–æ–Ω–≤–µ—Ä—Å–∏—é**.


**–ó–∞–¥–∞—á–∞ 2.4.** –ê–Ω–∞–ª–∏–∑ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö.
- –ù–∞ –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ ‚Äî —Å—Ä–µ–¥–Ω–µ–π —Å—É–º–º—ã –≤—Å–µ—Ö –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –£—á–∏—Ç—ã–≤–∞–π—Ç–µ –≤—Å–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–æ–≤–∞–ª–∏ —Å —Å–µ—Ä–≤–∏—Å–æ–º. –≠—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –∑–∞—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞—Ç—å –±–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –º–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏–∏.
- –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ, –≥–¥–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ. –ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Å–¥–µ–ª–∞–π—Ç–µ –∫—Ä–∞—Ç–∫–∏–π –≤—ã–≤–æ–¥.

````python
# üìå –°—É–º–º–∞—Ä–Ω—ã–π –¥–µ–ø–æ–∑–∏—Ç –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é
user_deposit_totals = (
    hist_data[hist_data['amount'].notna()]
    .groupby('user_id')['amount']
    .sum()
    .reset_index()
)

# üß© –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –±–µ–∑ –¥–µ–ø–æ–∑–∏—Ç–æ–≤
all_users = pd.DataFrame(hist_data['user_id'].unique(), columns=['user_id'])
user_deposit_totals = all_users.merge(user_deposit_totals, on='user_id', how='left')
user_deposit_totals['amount'] = user_deposit_totals['amount'].fillna(0)

# üìà –†–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–π —Å—É–º–º—ã –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
average_deposit_per_user = user_deposit_totals['amount'].mean()
std_deposit_per_user = user_deposit_totals['amount'].std()

# üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(10, 5))
sns.histplot(user_deposit_totals['amount'], bins=50, kde=True)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—É–º–º—ã –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è')
plt.xlabel('–°—É–º–º–∞ –¥–µ–ø–æ–∑–∏—Ç–æ–≤')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
plt.tight_layout()
plt.show()

# üßæ –í—ã–≤–æ–¥
print(f'–°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –≤—Å–µ—Ö –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {average_deposit_per_user:.2f}')
print(f'–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_deposit_per_user:.2f}')
````

- –°–¥–≤–∏–≥ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–ª–µ–≤–æ ‚Äî –ø—Ä–µ–æ–±–ª–∞–¥–∞–Ω–∏–µ –Ω—É–ª–µ–≤—ã—Ö –∏ –Ω–∏–∑–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π;
- –ü–∏–∫ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ **100‚Äì130**, —á—Ç–æ –æ—Ç—Ä–∞–∂–∞–µ—Ç —Ç–∏–ø–∏—á–Ω—É—é —Å—É–º–º—É –æ–¥–Ω–æ–≥–æ –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–æ–ø–æ–ª–Ω–µ–Ω–∏–π;
- –î–ª–∏–Ω–Ω—ã–π –ø—Ä–∞–≤—ã–π —Ö–≤–æ—Å—Ç (–¥–æ 600+), —É–∫–∞–∑—ã–≤–∞—é—â–∏–π –Ω–∞ —É–∑–∫—É—é, –Ω–æ –≤–µ—Å–æ–º—É—é –≥—Ä—É–ø–ø—É –∞–∫—Ç–∏–≤–Ω—ã—Ö –∏–Ω–≤–µ—Å—Ç–æ—Ä–æ–≤.

### 3. –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ê/–í —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞

````python
# üëÅ –ü–µ—Ä–≤–∏—á–Ω—ã–π –ø—Ä–æ—Å–º–æ—Ç—Ä
display(abt_data.head())
display(abt_data.info())
display(abt_data.isna().sum())
````

````python
# üïí –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞—Ç—ã –≤ —Ñ–æ—Ä–º–∞—Ç datetime
abt_data['first_ts'] = pd.to_datetime(abt_data['first_ts'])
abt_data['first_dt'] = pd.to_datetime(abt_data['first_dt'])
abt_data['event_ts'] = pd.to_datetime(abt_data['event_ts'])

# üî¢ –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—É–º–º—ã –¥–µ–ø–æ–∑–∏—Ç–æ–≤
abt_data['amount'] = pd.to_numeric(abt_data['amount'], errors='coerce')

# üßº –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–ø—É—Å–∫–æ–≤
abt_data['asset'] = abt_data['asset'].fillna('unknown')
abt_data['risk_level'] = abt_data['risk_level'].fillna('unknown')

# üßπ –£–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
abt_data = abt_data.drop_duplicates()
````

  - `amount` –∑–∞–ø–æ–ª–Ω–µ–Ω –¥–ª—è **7 843 —Å—Ç—Ä–æ–∫** ‚Äî —ç—Ç–æ —Å–æ–±—ã—Ç–∏—è —Å –¥–µ–ø–æ–∑–∏—Ç–∞–º–∏;

````python
# üìã –°—á–∏—Ç–∞–µ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ
users_by_group = (
    abt_data[['user_id', 'group']]
    .drop_duplicates()
    .groupby('group')
    .agg(users_count=('user_id', 'nunique'))
    .reset_index()
)

# üìà –°—Ç–æ–ª–±—á–∞—Ç–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
plt.figure(figsize=(6, 4))
sns.barplot(data=users_by_group, x='group', y='users_count')
plt.title(' –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –≥—Ä—É–ø–ø–∞–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞')
plt.xlabel('–ì—Ä—É–ø–ø–∞')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
plt.tight_layout()
plt.show()

# üìã –í—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã
users_by_group
````

* –°–ø–ª–∏—Ç –ø–æ–ª—É—á–∏–ª—Å—è **–±–ª–∏–∑–∫–∏–º –∫ 50/50**, –Ω–æ —Å –Ω–µ–±–æ–ª—å—à–æ–π —Ä–∞–∑–Ω–∏—Ü–µ–π –≤ —Ä–∞–∑–º–µ—Ä–µ: –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–∞—è –≥—Ä—É–ø–ø–∞ –Ω–∞ **~6% –±–æ–ª—å—à–µ**.
* –î–ª—è —Ü–µ–ª–µ–π —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ —Ç–∞–∫–∞—è —Ä–∞–∑–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∞ –¥–æ–ø—É—Å—Ç–∏–º–∞, –æ—Å–æ–±–µ–Ω–Ω–æ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É—é—â–∏—Ö –º–µ—Ç—Ä–∏–∫ –∏ –±—É—Ç—Å—Ç—Ä–∞–ø–∞ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏—Ö —ç—Ç–∞–ø–∞—Ö –∞–Ω–∞–ª–∏–∑–∞.

üìå –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –≥—Ä—É–ø–ø–∞–º –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∏ –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç—å –∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é –∫–ª—é—á–µ–≤—ã—Ö –º–µ—Ç—Ä–∏–∫.


**–ó–∞–¥–∞—á–∞ 3.2.** –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–æ—Ä–æ–Ω–æ–∫ —Å–æ–±—ã—Ç–∏–π.
- –°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –ø—É—Ç—å –Ω–æ–≤—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö.
- –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫—É—é –∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—É—é –≤–æ—Ä–æ–Ω–∫–∏, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é –∏ —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π –≤—ã–≤–æ–¥.

````python
# üìå –ü–æ–ª–Ω—ã–π –ø—É—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
funnel_steps = [
    'install / open_web',
    'introduction',
    'main_page',
    'onboarding_complete',
    'registration',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# –û—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º —Å–æ–±—ã—Ç–∏—è –ø–æ —ç—Ç–∏–º —à–∞–≥–∞–º
abt_funnel = abt_data[abt_data['event_name'].isin(funnel_steps)].copy()

# –£—Å—Ç–∞–Ω–æ–≤–∏–º –ø–æ—Ä—è–¥–æ–∫
abt_funnel['event_name'] = pd.Categorical(abt_funnel['event_name'], categories=funnel_steps, ordered=True)

# –£–¥–∞–ª–∏–º –¥—É–±–ª–∏–∫–∞—Ç—ã
abt_funnel_unique = abt_funnel.drop_duplicates(subset=['user_id', 'event_name', 'group'])

# üìä –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –ø–æ –≥—Ä—É–ø–ø–∞–º
funnel_by_group = (
    abt_funnel_unique
    .groupby(['group', 'event_name'])['user_id']
    .nunique()
    .reset_index(name='users_count')
)

# üìà –ü–∏–≤–æ—Ç-—Ç–∞–±–ª–∏—Ü–∞ –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏–π
funnel_pivot = funnel_by_group.pivot(index='event_name', columns='group', values='users_count').reset_index()

# üìâ –†–∞—Å—á—ë—Ç –ø–æ—à–∞–≥–æ–≤–æ–π –∏ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã
for grp in ['control', 'test']:
    funnel_pivot[f'{grp}_step_cr'] = funnel_pivot[grp] / funnel_pivot[grp].shift(1)
    funnel_pivot[f'{grp}_total_cr'] = funnel_pivot[grp] / funnel_pivot[grp].iloc[0]

# üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
plt.figure(figsize=(12, 5))
sns.barplot(data=funnel_by_group, x='event_name', y='users_count', hue='group')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —à–∞–≥–∞–º –≤–æ—Ä–æ–Ω–∫–∏ (–ø–æ –≥—Ä—É–ø–ø–∞–º)')
plt.xlabel('–°–æ–±—ã—Ç–∏–µ')
plt.ylabel('–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ—à–∞–≥–æ–≤–æ–π –∫–æ–Ω–≤–µ—Ä—Å–∏–∏
cr_plot = funnel_pivot.melt(
    id_vars='event_name',
    value_vars=['control_step_cr', 'test_step_cr'],
    var_name='group_step',
    value_name='conversion_rate'
)
cr_plot['group'] = cr_plot['group_step'].str.extract(r'(control|test)')

plt.figure(figsize=(12, 5))
sns.barplot(data=cr_plot, x='event_name', y='conversion_rate', hue='group')
plt.title('–ü–æ—à–∞–≥–æ–≤–∞—è –∫–æ–Ω–≤–µ—Ä—Å–∏—è (step conversion rate) –ø–æ –≥—Ä—É–ø–ø–∞–º')
plt.xlabel('–®–∞–≥ –≤–æ—Ä–æ–Ω–∫–∏')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# üìã –¢–∞–±–ª–∏—Ü–∞ —Å –∏—Ç–æ–≥–∞–º–∏
funnel_pivot
````

# üîç –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π –≤—ã–≤–æ–¥: –≤–æ—Ä–æ–Ω–∫–∞ —Å–æ–±—ã—Ç–∏–π –ø–æ –≥—Ä—É–ø–ø–∞–º

* **–ü–µ—Ä–≤—ã–µ —à–∞–≥–∏ (–¥–æ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏)** –ø—Ä–æ—Ö–æ–¥—è—Ç –ø–æ—á—Ç–∏ –≤—Å–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≤ –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö:

  - `onboarding_complete`: **Control** ‚Äî 83.8%, **Test** ‚Äî 75.8% ‚Üí –≤–æ–∑–º–æ–∂–Ω—ã–π —ç—Ñ—Ñ–µ–∫—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç–∏/—É—Å—Ç–∞–ª–æ—Å—Ç–∏ –æ—Ç –Ω–æ–≤–æ–≥–æ –æ–Ω–±–æ—Ä–¥–∏–Ω–≥–∞.

* **–†–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è** –ø—Ä–æ—Ö–æ–¥–∏—Ç **—á—É—Ç—å –≤—ã—à–µ –¥–æ–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π** –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø–µ:
  - **Control** ‚Äî 93.1%,

  ‚û§ –≠—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –æ–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π –æ–Ω–±–æ—Ä–¥–∏–Ω–≥ **–Ω–µ –º–µ—à–∞–µ—Ç –∑–∞–≤–µ—Ä—à–µ–Ω–∏—é —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏**.

* **–ö–æ–Ω–≤–µ—Ä—Å–∏—è –∏–∑ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –≤ –ø–µ—Ä–≤—ã–π –¥–µ–ø–æ–∑–∏—Ç**:

* **–ö–ª—é—á–µ–≤–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞**:
  - `asset_purchase`:  
    **Control** ‚Äî 36.6%,  
    **Test** ‚Äî 43.3%  
  - `second_deposit`:  
    **Control** ‚Äî 13.5%,  
    **Test** ‚Äî 21.3%  
  ‚û§ –ü—Ä–∏—Ä–æ—Å—Ç **–ø–æ—á—Ç–∏ –Ω–∞ 7.8 –ø.–ø.** –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç –Ω–æ–≤–æ–π —Ñ–∏—á–∏ –Ω–∞ –≤–æ–≤–ª–µ—á–µ–Ω–∏–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∏–Ω–≤–µ—Å—Ç–æ—Ä–æ–≤.

---

## üìà –ê–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:

* –ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –º–µ–Ω—å—à–∏–π –∏—Å—Ö–æ–¥–Ω—ã–π —Ä–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã (**4568 vs 4847**), –¥–æ –≤—Ç–æ—Ä–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞ –¥–æ—à–ª–æ:
  - **Control** ‚Äî 654 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π  
  - **Test** ‚Äî 974 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π  
  ‚û§ –≠—Ç–æ **–Ω–∞ 49% –±–æ–ª—å—à–µ**.

---

## üß© –í—ã–≤–æ–¥:

–û–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π –æ–Ω–±–æ—Ä–¥–∏–Ω–≥:
* **–Ω–µ –º–µ—à–∞–µ—Ç –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—é —Ä–∞–Ω–Ω–∏—Ö —ç—Ç–∞–ø–æ–≤** (—Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏—è –∏ –ø–µ—Ä–≤—ã–π –¥–µ–ø–æ–∑–∏—Ç),
* **—É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ—Å—Ç—å –ø–æ—Å–ª–µ –ø–æ–ø–æ–ª–Ω–µ–Ω–∏—è —Å—á—ë—Ç–∞**: –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –∞–∫—Ç–∏–≤–Ω–µ–µ —Å–æ–≤–µ—Ä—à–∞—é—Ç –¥–µ–π—Å—Ç–≤–∏—è –∏ –¥–µ–ª–∞—é—Ç –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –¥–µ–ø–æ–∑–∏—Ç—ã,
* –æ—Å–æ–±–µ–Ω–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ **—Å—Ä–µ–¥–Ω–∏–π –∏ –≤—ã—Å–æ–∫–∏–π —Å–µ–≥–º–µ–Ω—Ç—ã –∞—É–¥–∏—Ç–æ—Ä–∏–∏**, —á—Ç–æ —Å–æ–≥–ª–∞—Å—É–µ—Ç—Å—è —Å –∞–Ω–∞–ª–∏–∑–æ–º –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π.

üìå –§–∏—á–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –∫–∞–∫ —É—Å–∏–ª–∏—Ç–µ–ª—å –¥–ª—è —Ç–µ—Ö, –∫—Ç–æ –¥–æ—à—ë–ª –¥–æ –∫–ª—é—á–µ–≤—ã—Ö –¥–µ–π—Å—Ç–≤–∏–π ‚Äî –∏–º–µ–Ω–Ω–æ –≤ —ç—Ç–æ–º —Å–µ–≥–º–µ–Ω—Ç–µ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –ø—Ä–∏—Ä–æ—Å—Ç –≤ –º–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏–∏.


**–ó–∞–¥–∞—á–∞ 3.3.** –í–ª–∏—è–Ω–∏–µ –Ω–æ–≤–æ–π —Ñ–∏—á–∏ –Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏—é –≤–æ –≤—Ç–æ—Ä–æ–π –¥–µ–ø–æ–∑–∏—Ç —Å —É—á—ë—Ç–æ–º —É—Ä–æ–≤–Ω—è —Ä–∏—Å–∫–∞ –∫—É–ø–ª–µ–Ω–Ω–æ–≥–æ –∞–∫—Ç–∏–≤–∞.
- –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ, –∫–∞–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏—è —Ä–∏—Å–∫–∞ –∫—É–ø–ª–µ–Ω–Ω–æ–≥–æ –∞–∫—Ç–∏–≤–∞ –≤–ª–∏—è–µ—Ç –Ω–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–∫—Ä—ã—Ç–∏—è –≤—Ç–æ—Ä–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö.
- –°—Ä–∞–≤–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏ –æ—Ü–µ–Ω–∏—Ç–µ, –µ—Å—Ç—å –ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏.

````python
# –¢–æ–ª—å–∫–æ –≤–∞–ª–∏–¥–Ω—ã–µ —É—Ä–æ–≤–Ω–∏ —Ä–∏—Å–∫–∞
valid_assets = abt_data[
    abt_data['risk_level'].isin(['low', 'medium', 'high'])
].copy()

# –ü–µ—Ä–≤—ã–π –∞–∫—Ç–∏–≤ –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
first_assets = (
    valid_assets.sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()
    [['user_id', 'group', 'risk_level']]
)
# –ú–∞—Å—Å–∏–≤ user_id —Å —Å–æ–±—ã—Ç–∏–µ–º second_deposit
second_deposit_users = abt_data[
    abt_data['event_name'] == 'second_deposit'
]['user_id'].unique()

first_assets['second_deposit'] = first_assets['user_id'].isin(second_deposit_users)
# –ê–≥—Ä–µ–≥–∏—Ä—É–µ–º
risk_conv_grouped = (
    first_assets
    .groupby(['group', 'risk_level'])
    .agg(users_total=('user_id', 'count'),
         converted=('second_deposit', 'sum'))
    .reset_index()
)

# –ö–æ–Ω–≤–µ—Ä—Å–∏—è
risk_conv_grouped['conversion_rate'] = risk_conv_grouped['converted'] / risk_conv_grouped['users_total']
plt.figure(figsize=(10, 5))
sns.barplot(data=risk_conv_grouped, x='risk_level', y='conversion_rate', hue='group')
plt.title('–ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤–æ –≤—Ç–æ—Ä–æ–π –¥–µ–ø–æ–∑–∏—Ç –ø–æ —É—Ä–æ–≤–Ω—é —Ä–∏—Å–∫–∞ –∏ –≥—Ä—É–ø–ø–µ')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å —Ä–∏—Å–∫–∞ –ø–µ—Ä–≤–æ–≥–æ –∞–∫—Ç–∏–≤–∞')
plt.tight_layout()
plt.show()
risk_conv_grouped
````

````python
from statsmodels.stats.proportion import proportions_ztest

# –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –≥–∏–ø–æ—Ç–µ–∑ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ –∫–∞–∂–¥–æ–º—É —É—Ä–æ–≤–Ω—é —Ä–∏—Å–∫–∞
results = []

for risk in ['low', 'medium', 'high']:
    group_data = risk_conv_grouped[risk_conv_grouped['risk_level'] == risk]

    control_conv = group_data[group_data['group'] == 'control']
    test_conv = group_data[group_data['group'] == 'test']

    successes = [test_conv['converted'].values[0], control_conv['converted'].values[0]]
    totals = [test_conv['users_total'].values[0], control_conv['users_total'].values[0]]

    stat, pval = proportions_ztest(successes, totals)
    results.append({'risk_level': risk, 'z_stat': stat, 'p_value': pval})

# –ò—Ç–æ–≥–æ–≤–∞—è —Ç–∞–±–ª–∏—Ü–∞
pd.DataFrame(results)
````

````python
from statsmodels.stats.proportion import proportions_ztest

# –î–∞–Ω–Ω—ã–µ –ø–æ –≥—Ä—É–ø–ø–∞–º
reg_control = registered_users[registered_users['group'] == 'control']
reg_test = registered_users[registered_users['group'] == 'test']

successes = [reg_test['has_deposit'].sum(), reg_control['has_deposit'].sum()]
totals = [reg_test.shape[0], reg_control.shape[0]]

# z-—Ç–µ—Å—Ç –ø—Ä–æ–ø–æ—Ä—Ü–∏–π
stat, pval_cr_reg = proportions_ztest(successes, totals)

# –†–∞—Å—á—ë—Ç –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –∏ –¥–µ–ª—å—Ç
cr_control = successes[1] / totals[1]
cr_test = successes[0] / totals[0]
delta_abs = cr_test - cr_control
delta_rel = (delta_abs / cr_control) * 100

# –¢–∞–±–ª–∏—Ü–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
import pandas as pd
cr_reg_result = pd.DataFrame({
    'group': ['control', 'test'],
    'cr_registration_to_deposit': [cr_control, cr_test],
})
delta_row = pd.DataFrame({
    'group': ['—Ä–∞–∑–Ω–∏—Ü–∞ (test - control)'],
    'cr_registration_to_deposit': [f"{delta_abs:.4f} ({delta_rel:.2f}%)"],
})
cr_reg_result = pd.concat([cr_reg_result, delta_row], ignore_index=True)

cr_reg_result, pval_cr_reg
````

üìå **–í—ã–≤–æ–¥:** –°–Ω–∏–∂–µ–Ω–∏–µ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø–µ **–Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–º**. –û–¥–Ω–∞–∫–æ —Ç—Ä–µ–Ω–¥ –≤ —Å—Ç–æ—Ä–æ–Ω—É —Å–Ω–∏–∂–µ–Ω–∏—è –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ **–Ω–µ–≥–∞—Ç–∏–≤–Ω–æ–º –≤–ª–∏—è–Ω–∏–∏ —Ñ–∏—á–∏ –Ω–∞ –º–µ–Ω–µ–µ —É–≤–µ—Ä–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π**.


### –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ 1: –ö–æ–Ω–≤–µ—Ä—Å–∏—è –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞ –≤–æ –≤—Ç–æ—Ä–æ–π (`cr_deposit_to_second`)

**H0 (–Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞):** –ö–æ–Ω–≤–µ—Ä—Å–∏—è –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞ –≤–æ –≤—Ç–æ—Ä–æ–π –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö.  
**H1 (–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞):** –ö–æ–Ω–≤–µ—Ä—Å–∏–∏ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏.


````python
# –î–∞–Ω–Ω—ã–µ –ø–æ –≥—Ä—É–ø–ø–∞–º
conv_control = depositers[depositers['group'] == 'control']
conv_test = depositers[depositers['group'] == 'test']

successes = [conv_test['second'].sum(), conv_control['second'].sum()]
totals = [conv_test.shape[0], conv_control.shape[0]]

# z-—Ç–µ—Å—Ç –ø—Ä–æ–ø–æ—Ä—Ü–∏–π
stat, pval_cr_second = proportions_ztest(successes, totals)

# –†–∞—Å—á—ë—Ç –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –∏ –¥–µ–ª—å—Ç
cr_control = successes[1] / totals[1]
cr_test = successes[0] / totals[0]
delta_abs = cr_test - cr_control
delta_rel = (delta_abs / cr_control) * 100

# –¢–∞–±–ª–∏—Ü–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
cr_second_result = pd.DataFrame({
    'group': ['control', 'test'],
    'cr_deposit_to_second': [cr_control, cr_test],
})
delta_row = pd.DataFrame({
    'group': ['—Ä–∞–∑–Ω–∏—Ü–∞ (test - control)'],
    'cr_deposit_to_second': [f"{delta_abs:.4f} ({delta_rel:.2f}%)"],
})
cr_second_result = pd.concat([cr_second_result, delta_row], ignore_index=True)

cr_second_result, pval_cr_second
````

### üìä –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ 1 ‚Äî –ö–æ–Ω–≤–µ—Ä—Å–∏—è –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞ –≤–æ –≤—Ç–æ—Ä–æ–π

| –†–∞–∑–Ω–∏—Ü–∞  | +12.35 –ø.–ø. (+60.95%) |

**p-value < 0.001** ‚Üí –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ **–æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è**.

üìå **–í—ã–≤–æ–¥:** –ù–æ–≤–∞—è —Ñ–∏—á–∞ **–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—Ç–æ—Ä–æ–≥–æ –¥–µ–ø–æ–∑–∏—Ç–∞**. –≠—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–π –∏ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –≤–∞–∂–Ω—ã–π —ç—Ñ—Ñ–µ–∫—Ç, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π –Ω–∞ —Ä–æ—Å—Ç –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ—Å—Ç–∏.


### –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ 2: –°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –≤—Å–µ—Ö –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, —Å–æ–≤–µ—Ä—à–∏–≤—à–µ–≥–æ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –ø–ª–∞—Ç—ë–∂ (`avg_deposit_positive`)

**H0 (–Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞):** –°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –¥–µ–ø–æ–∑–∏—Ç–æ–≤ —É –ø–ª–∞—Ç—è—â–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö.  
**H1 (–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞):** –°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –¥–µ–ø–æ–∑–∏—Ç–æ–≤ —É –ø–ª–∞—Ç—è—â–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏.


````python
from scipy.stats import ttest_ind

# –í—ã–±–æ—Ä–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º
pos_control = deposit_only[deposit_only['group'] == 'control']['amount']
pos_test = deposit_only[deposit_only['group'] == 'test']['amount']

# t-—Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞
stat, pval_positive_avg = ttest_ind(pos_test, pos_control, equal_var=False)

# –°—Ä–µ–¥–Ω–∏–µ –∏ –¥–µ–ª—å—Ç—ã
avg_control = pos_control.mean()
avg_test = pos_test.mean()
delta_abs = avg_test - avg_control
delta_rel = (delta_abs / avg_control) * 100

# –¢–∞–±–ª–∏—Ü–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
avg_positive_result = pd.DataFrame({
    'group': ['control', 'test'],
    'avg_deposit_positive': [avg_control, avg_test],
})
delta_row = pd.DataFrame({
    'group': ['—Ä–∞–∑–Ω–∏—Ü–∞ (test - control)'],
    'avg_deposit_positive': [f"{delta_abs:.2f} ({delta_rel:.2f}%)"],
})
avg_positive_result = pd.concat([avg_positive_result, delta_row], ignore_index=True)

avg_positive_result, pval_positive_avg
````

### üìä –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ 2 ‚Äî –°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –¥–µ–ø–æ–∑–∏—Ç–æ–≤ —Å—Ä–µ–¥–∏ –ø–ª–∞—Ç—è—â–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

**p-value = 0.040** ‚Üí –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ **–æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è**.

üìå **–í—ã–≤–æ–¥:** –¢–µ—Å—Ç–æ–≤–∞—è –≥—Ä—É–ø–ø–∞ —Å—Ä–µ–¥–∏ –ø–ª–∞—Ç—è—â–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç **—Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–π —Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–π —Å—É–º–º—ã –¥–µ–ø–æ–∑–∏—Ç–æ–≤**. –≠—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Ñ–∏—á–∞ —É—Å–∏–ª–∏–≤–∞–µ—Ç –ø–æ–≤–µ–¥–µ–Ω–∏–µ –±–æ–ª–µ–µ –≤–æ–≤–ª–µ—á—ë–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.


### –ê–Ω–∞–ª–∏–∑ –¥–∏–Ω–∞–º–∏–∫–∏ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ –ø–æ –¥–Ω—è–º

**–¶–µ–ª—å:** –ò–∑—É—á–∏—Ç—å –Ω–∞–∫–æ–ø–ª–µ–Ω–Ω—É—é —Å—Ä–µ–¥–Ω—é—é —Å—É–º–º—É –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ –¥–Ω—è–º –∏ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å p-value, —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, –∫–æ–≥–¥–∞ —ç—Ñ—Ñ–µ–∫—Ç —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —É—Å—Ç–æ–π—á–∏–≤—ã–º –∏ –∑–Ω–∞—á–∏–º—ã–º.


````python
# –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞: —Å—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ –¥–Ω—è–º
cum_daily = (
    daily_deposit_full
    .groupby(['event_date', 'group'])['amount']
    .mean()
    .reset_index()
)

# üìà –ì—Ä–∞—Ñ–∏–∫ –¥–∏–Ω–∞–º–∏–∫–∏
plt.figure(figsize=(12, 5))
for group in cum_daily['group'].unique():
    subset = cum_daily[cum_daily['group'] == group]
    plt.plot(subset['event_date'], subset['amount'], label=group)
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ —Å—Ä–µ–¥–Ω–µ–π —Å—É–º–º—ã –¥–µ–ø–æ–∑–∏—Ç–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ –¥–Ω—è–º')
plt.xlabel('–î–∞—Ç–∞')
plt.ylabel('–°—Ä–µ–¥–Ω—è—è —Å—É–º–º–∞ –¥–µ–ø–æ–∑–∏—Ç–æ–≤')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# üìã –¢–∞–±–ª–∏—Ü–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
cum_daily_pivot = cum_daily.pivot(index='event_date', columns='group', values='amount').reset_index()
cum_daily_pivot
````

````python
# –†–∞—Å—á—ë—Ç p-value –ø–æ –¥–Ω—è–º
pvalue_timeline = []

for day in sorted(daily_deposit_full['event_date'].dropna().unique()):
    snapshot = daily_deposit_full[daily_deposit_full['event_date'] <= day]

    ctrl = snapshot[snapshot['group'] == 'control'].groupby('user_id')['amount'].sum()
    test = snapshot[snapshot['group'] == 'test'].groupby('user_id')['amount'].sum()

    stat = ttest_ind(test, ctrl, equal_var=False)
    pvalue_timeline.append({'date': day, 'p_value': stat.pvalue})

pvalue_df = pd.DataFrame(pvalue_timeline)

# üìà –ì—Ä–∞—Ñ–∏–∫ p-value
plt.figure(figsize=(12, 4))
plt.plot(pvalue_df['date'], pvalue_df['p_value'], marker='o')
plt.axhline(0.05, color='red', linestyle='--', label='Œ± = 0.05')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ p-value –ø–æ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–µ')
plt.xlabel('–î–∞—Ç–∞')
plt.ylabel('p-value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# üìã –¢–∞–±–ª–∏—Ü–∞
pvalue_df
````

```python
# –ü—Ä–∏–º–µ—Ä –±—É—Ç—Å—Ç—Ä–∞–ø-—Ç–µ—Å—Ç–∞
import numpy as np

# –§–∏–∫—Å–∏—Ä—É–µ–º —Å–∏–¥ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
np.random.seed(341)

# –î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: –∫–æ–Ω—Ç—Ä–æ–ª—å –∏ —Ç–µ—Å—Ç –ø–æ 1000 –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
# —Å–æ —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ 2.2 –∏ 1.7
control = np.random.exponential(scale=2.2, size=1000)
test = np.random.exponential(scale=1.7, size=1000)

n_iterations = 1000
boot_diffs = []

# –ë—É—Ç—Å—Ç—Ä–∞–ø-—Ä–∞–∑–Ω–∏—Ü–∞ 95-—Ö –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π –º–µ–∂–¥—É —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø–∞–º–∏
for i in range(n_iterations):
    boot_control = np.random.choice(control, size=len(control), replace=True)
    boot_test = np.random.choice(test, size=len(test), replace=True)

    # –î–ª—è —Ä–∞—Å—á—ë—Ç–∞ 95 –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è –∏—Å–ø–æ–ª—å–∑—É–µ–º .percentile() –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ numpy
    control_95p = np.percentile(boot_control, 95)
    test_95p = np.percentile(boot_test, 95)

    # –ù–∞—Ö–æ–¥–∏–º —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø
    boot_diffs.append(test_95p - control_95p)

# –°–æ–∑–¥–∞—ë–º –º–∞—Å—Å–∏–≤ —Ä–∞–∑–Ω–∏—Ü –º–µ–∂–¥—É 95-–º–∏ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π –≥—Ä—É–ø–ø
boot_diffs = np.array(boot_diffs)

diff_ci = np.percentile(boot_diffs, [2.5, 97.5])
```

````python
# –¢–æ–ª—å–∫–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏–µ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –¥–µ–ø–æ–∑–∏—Ç
paid_users = deposit_all_users[deposit_all_users['amount'] > 0].copy()

# –†–∞–∑–¥–µ–ª–∏–º –ø–æ –≥—Ä—É–ø–ø–∞–º
control_group = paid_users[paid_users['group'] == 'control']['amount'].values
test_group = paid_users[paid_users['group'] == 'test']['amount'].values
import numpy as np

# –§–∏–∫—Å–∏—Ä—É–µ–º —Å–∏–¥
np.random.seed(42)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
n_iterations = 1000
percentiles = [25, 50, 75]
bootstrap_results = {}

for p in percentiles:
    boot_diffs = []
    for _ in range(n_iterations):
        control_sample = np.random.choice(control_group, size=len(control_group), replace=True)
        test_sample = np.random.choice(test_group, size=len(test_group), replace=True)

        diff = np.percentile(test_sample, p) - np.percentile(control_sample, p)
        boot_diffs.append(diff)

    boot_diffs = np.array(boot_diffs)
    ci = np.percentile(boot_diffs, [2.5, 97.5])

    bootstrap_results[p] = {
        'control_percentile': np.percentile(control_group, p),
        'test_percentile': np.percentile(test_group, p),
        'mean_diff': np.mean(boot_diffs),
        'ci_low': ci[0],
        'ci_high': ci[1],
        'distribution': boot_diffs
    }
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))

for i, p in enumerate(percentiles):
    plt.subplot(1, 3, i + 1)
    plt.hist(bootstrap_results[p]['distribution'], bins=50, alpha=0.7)
    plt.axvline(bootstrap_results[p]['ci_low'], color='red', linestyle='--')
    plt.axvline(bootstrap_results[p]['ci_high'], color='red', linestyle='--')
    plt.axvline(0, color='black')
    plt.title(f'–†–∞–∑–Ω–∏—Ü–∞ {p}-–≥–æ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è\n95% CI: [{bootstrap_results[p]["ci_low"]:.2f}, {bootstrap_results[p]["ci_high"]:.2f}]')
    plt.xlabel('–†–∞–∑–Ω–∏—Ü–∞ (test - control)')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

plt.tight_layout()
plt.show()
import pandas as pd

summary = pd.DataFrame([
    {
        'percentile': p,
        'control': bootstrap_results[p]['control_percentile'],
        'test': bootstrap_results[p]['test_percentile'],
        'mean_diff': bootstrap_results[p]['mean_diff'],
        'ci_low': bootstrap_results[p]['ci_low'],
        'ci_high': bootstrap_results[p]['ci_high']
    }
    for p in percentiles
])

summary
````
