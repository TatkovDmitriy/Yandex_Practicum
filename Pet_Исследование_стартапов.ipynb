```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤
print("–ò—Å—Ö–æ–¥–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –ø–µ—Ä–µ–¥ –æ–±—Ä–∞–±–æ—Ç–∫–æ–π:")
print(company_and_rounds.columns)

# –£–±–∏—Ä–∞–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
company_and_rounds.columns = company_and_rounds.columns.str.strip()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –ø—Ä–æ–±–µ–ª–æ–≤
duplicate_columns = company_and_rounds.columns[company_and_rounds.columns.duplicated(keep=False)]
if duplicate_columns.any():
    print("\n‚ö†Ô∏è –ù–∞–π–¥–µ–Ω—ã –¥—É–±–ª–∏—Ä—É—é—â–∏–µ—Å—è —Å—Ç–æ–ª–±—Ü—ã:", duplicate_columns.tolist())

# –ï—Å–ª–∏ –µ—Å—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã, —è–≤–Ω–æ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –æ–¥–∏–Ω –∏–∑ –Ω–∏—Ö
if 'company  ID' in company_and_rounds.columns and 'company  id' in company_and_rounds.columns:
    company_and_rounds.rename(columns={'company  ID': 'company_id_main', 'company  id': 'company_id_alt'}, inplace=True)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ snake_case
company_and_rounds.columns = (
    company_and_rounds.columns
    .str.replace(r'\s+', '_', regex=True)  # –ó–∞–º–µ–Ω—è–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–æ–±–µ–ª—ã –æ–¥–Ω–∏–º _
    .str.lower()  # –ü—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
)

# –£–¥–∞–ª—è–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
company_and_rounds = company_and_rounds.loc[:, ~company_and_rounds.columns.duplicated()]

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è
print("\n–û–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏:")
print(company_and_rounds.columns)

# –í—ã–≤–æ–¥–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Å–µ—Ç–µ
print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ datatypes –¥–ª—è company_and_rounds:")
print(company_and_rounds.info(), "\n")
```

```python
def analyze_datasets(dfs, names):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —Å–ø–∏—Å–∫–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤.

    Parameters:
    dfs (list of pd.DataFrame): –°–ø–∏—Å–æ–∫ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.
    names (list of str): –°–ø–∏—Å–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π —Ñ–∞–π–ª–æ–≤ (–¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è).
    """

    for df, name in zip(dfs, names):
        print(f"\nüîç –ê–Ω–∞–ª–∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞: {name}")
        print(f"üìä –†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å: {df.shape[0]} —Å—Ç—Ä–æ–∫, {df.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤")
        print("-" * 50)

        # –ü–æ–¥—Å—á—ë—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤
        missing_percent = df.isna().mean() * 100
        missing_summary = missing_percent[missing_percent > 0].sort_values(ascending=False)
        if not missing_summary.empty:
            print("‚ö†Ô∏è –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–≤ %):")
            display(missing_summary.to_frame().T)
        else:
            print("‚úÖ –ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ—Ç.")

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
        duplicate_percent = df.duplicated().mean() * 100
        if duplicate_percent > 0:
            print(f"‚ö†Ô∏è –î—É–±–ª–∏–∫–∞—Ç—ã: {duplicate_percent:.2f}% –∑–∞–ø–∏—Å–µ–π.")
        else:
            print("‚úÖ –î—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–µ—Ç.")

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—É—Å—Ç—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
        empty_columns = df.columns[df.isna().all()]
        if empty_columns.any():
            print(f"‚ö†Ô∏è –ü–æ–ª–Ω–æ—Å—Ç—å—é –ø—É—Å—Ç—ã–µ —Å—Ç–æ–ª–±—Ü—ã: {list(empty_columns)}")

        # –í—ã–≤–æ–¥ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–∏–º–µ—Ä–æ–≤
        print("\nüìå –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–∏–º–µ—Ä—ã –∑–Ω–∞—á–µ–Ω–∏–π:")
        sample_data = df.iloc[0:1].T  # –ë–µ—Ä–µ–º –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É –≤ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤–∏–¥–µ
        sample_data.columns = ["–ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö"]
        sample_data.insert(0, "–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö", df.dtypes)
        display(sample_data)

        print("=" * 80)

# –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—Å–µ—Ö –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
analyze_datasets(
    dfs=[acquisition, company_and_rounds, degrees, education, fund, investment, people],
    names=["acquisition", "company_and_rounds", "degrees", "education", "fund", "investment", "people"]
)
```

  * —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤, –≤—ã–¥–µ–ª—è–µ–º—ã–π –≤ —Ä–∞–º–∫–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞—É–Ω–¥–∞;
  * –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞—É–Ω–¥–æ–≤ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è –∑–∞ —ç—Ç–æ—Ç –≥–æ–¥.

* –û—Å—Ç–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–æ–ª—å–∫–æ –¥–ª—è —Ç–µ—Ö –ª–µ—Ç, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –±–æ–ª–µ–µ —á–µ–º 50 —Ä–∞—É–Ω–¥–∞—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è.

* –ù–∞ –æ—Å–Ω–æ–≤–µ –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è —Ç–∞–±–ª–∏—Ü—ã –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –æ—Ç—Ä–∞–∂–∞—Ç—å –¥–∏–Ω–∞–º–∏–∫—É —Ç–∏–ø–∏—á–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ —Å—Ä–µ–¥—Å—Ç–≤, –∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–∞—Ä—Ç–∞–ø—ã –ø–æ–ª—É—á–∞–ª–∏ –≤ —Ä–∞–º–∫–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è.

–ù–∞ –æ—Å–Ω–æ–≤–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã:

* –í –∫–∞–∫–æ–º –≥–æ–¥—É —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—Ä–∞–Ω–Ω—ã—Ö –≤ —Ä–∞–º–∫–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Å—Ä–µ–¥—Å—Ç–≤ –±—ã–ª –º–∞–∫—Å–∏–º–∞–ª–µ–Ω?

* –ö–∞–∫–∞—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏—è –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ä–∞—É–Ω–¥–æ–≤ –∏ –≤—ã–¥–µ–ª—è–µ–º—ã—Ö –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∂–¥–æ–≥–æ —Ä–∞—É–Ω–¥–∞ —Å—Ä–µ–¥—Å—Ç–≤ –Ω–∞–±–ª—é–¥–∞–ª–∞—Å—å –≤ 2013 –≥–æ–¥—É?

```python
# –®–∞–≥ 1: –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ 'funded_at' –≤ datetime –∏ –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ –≥–æ–¥–∞
company_and_rounds['funded_at'] = pd.to_datetime(company_and_rounds['funded_at'], errors='coerce')

# –ü—Ä–æ–≤–µ—Ä–∏–º, –µ—Å—Ç—å –ª–∏ –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—Å–ª–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
print(company_and_rounds['funded_at'].isnull().sum(), "–∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ —É–¥–∞–ª–æ—Å—å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤ –¥–∞—Ç—É.")

# –ò–∑–≤–ª–µ–∫–∞–µ–º –≥–æ–¥, –µ—Å–ª–∏ 'funded_at' –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω –≤ datetime
company_and_rounds['funding_year'] = company_and_rounds['funded_at'].dt.year

# –£–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ 'raised_amount' –∏ 'funding_year'
company_and_rounds = company_and_rounds.dropna(subset=['raised_amount', 'funding_year'])

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
display(company_and_rounds.head())
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç–æ–ª–±–µ—Ü 'founded_at' –≤ datetime, –∑–∞–º–µ–Ω—è—è –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ NaT
company_and_rounds['founded_at'] = pd.to_datetime(company_and_rounds['founded_at'], errors='coerce')

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞ –∏ –Ω–∞–ª–∏—á–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
print(f"–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞ 'founded_at': {company_and_rounds['founded_at'].dtype}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ 'founded_at': {company_and_rounds['founded_at'].isnull().sum()}")

# –ï—Å–ª–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ, –º–æ–∂–Ω–æ –∏–∑–≤–ª–µ—á—å –≥–æ–¥ –∏–∑ 'founded_at' —Ç–∞–∫ –∂–µ, –∫–∞–∫ –∏ –∏–∑ 'funded_at'
company_and_rounds['founded_year'] = company_and_rounds['founded_at'].dt.year

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
display(company_and_rounds.head())
```

```python
display(company_and_rounds.info())
```

```python
# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ 'founded_year' –∑–Ω–∞—á–µ–Ω–∏–µ–º '–ù–µ —É–∫–∞–∑–∞–Ω–æ'
company_and_rounds['founded_year'].fillna('–ù–µ —É–∫–∞–∑–∞–Ω–æ', inplace=True)

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
display(company_and_rounds.head())
```

```python
display(company_and_rounds.info())
```

```python
# –®–∞–≥ 2: –°–æ—Å—Ç–∞–≤–ª—è–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –≥–æ–¥–∞–º
funding_by_year = company_and_rounds.groupby('funding_year').agg(
    typical_amount=('raised_amount', 'median'),  # —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤ (–º–µ–¥–∏–∞–Ω–∞)
    total_rounds=('raised_amount', 'size')      # –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞—É–Ω–¥–æ–≤
)

# –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –≥–æ–¥—ã —Å –±–æ–ª–µ–µ —á–µ–º 50 —Ä–∞—É–Ω–¥–∞–º–∏
funding_by_year = funding_by_year[funding_by_year['total_rounds'] > 50]

# –í—ã–≤–æ–¥–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
funding_by_year.head()
```

```python
# –®–∞–≥ 3: –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ —Ç–∞–±–ª–∏—Ü—ã —Å –≥–æ–¥–∞–º–∏ –≤ –º–∏–ª–ª–∏–æ–Ω–∞—Ö –¥–æ–ª–ª–∞—Ä–æ–≤
plt.figure(figsize=(12, 8))

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∏–Ω–¥–µ–∫—Å –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤
x = funding_by_year.index.to_numpy()  # –ì–æ–¥ (–æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤)
y = funding_by_year['typical_amount'].values / 1_000_000  # –¢–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤ –≤ –º–∏–ª–ª–∏–æ–Ω–∞—Ö –¥–æ–ª–ª–∞—Ä–æ–≤

# –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∏–ø–∏—á–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ —Å—Ä–µ–¥—Å—Ç–≤ (–º–µ–¥–∏–∞–Ω–∞) –ø–æ –≥–æ–¥–∞–º
plt.subplot(2, 1, 1)  # –í–µ—Ä—Ö–Ω—è—è —á–∞—Å—Ç—å –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.plot(x, y, marker='o', color='b', label='–¢–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ —Ç–∏–ø–∏—á–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ —Å—Ä–µ–¥—Å—Ç–≤ –ø–æ –≥–æ–¥–∞–º (–≤ –º–ª–Ω –¥–æ–ª–ª–∞—Ä–æ–≤)')
plt.xlabel('–ì–æ–¥')
plt.ylabel('–¢–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤ (–≤ –º–ª–Ω –¥–æ–ª–ª–∞—Ä–æ–≤)')
plt.grid(True)
plt.legend()

# –°–æ–∑–¥–∞–µ–º —Ç–∞–±–ª–∏—Ü—É —Å –≥–æ–¥–∞–º–∏ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º —Å—Ç–æ–ª–±—Ü–µ
table_data = funding_by_year.reset_index()  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∏–Ω–¥–µ–∫—Å –≤ —Å—Ç–æ–ª–±–µ—Ü
table_data['typical_amount'] = table_data['typical_amount']

# –û–∫—Ä—É–≥–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ–ª—å–∫–æ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö 'typical_amount' –∏ 'total_rounds'
table_data['typical_amount'] = table_data['typical_amount'].apply(lambda x: f"{x:,.0f}")
table_data['total_rounds'] = table_data['total_rounds'].apply(lambda x: f"{x:,.0f}")

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ç–∞–±–ª–∏—Ü—É –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º
plt.subplot(2, 1, 2)  # –ù–∏–∂–Ω—è—è —á–∞—Å—Ç—å –¥–ª—è —Ç–∞–±–ª–∏—Ü—ã
plt.table(cellText=table_data.values,
          colLabels=table_data.columns,
          cellLoc='center', loc='center')
plt.axis('off')  # –û—Ç–∫–ª—é—á–∞–µ–º –æ—Å–∏ –¥–ª—è —Ç–∞–±–ª–∏—Ü—ã

plt.tight_layout()  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ —Ä–∞–∑–º–µ—â–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ
plt.show()
```

```python
# –®–∞–≥ 4: –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã

# 1. –ì–æ–¥ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —Ç–∏–ø–∏—á–Ω—ã–º —Ä–∞–∑–º–µ—Ä–æ–º —Å—Ä–µ–¥—Å—Ç–≤
max_funding_year = funding_by_year['typical_amount'].idxmax()
max_funding_value = funding_by_year['typical_amount'].max()

# 2. –¢–µ–Ω–¥–µ–Ω—Ü–∏–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ä–∞—É–Ω–¥–æ–≤ –∏ —Å—Ä–µ–¥—Å—Ç–≤–∞–º –≤ 2013 –≥–æ–¥—É
funding_2013 = funding_by_year.loc[2013]
rounds_2013 = company_and_rounds[company_and_rounds['funding_year'] == 2013].shape[0]
typical_amount_2013 = funding_2013['typical_amount']

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤ –±—ã–ª –≤ –≥–æ–¥—É {max_funding_year} –∏ —Å–æ—Å—Ç–∞–≤–∏–ª {max_funding_value} –¥–æ–ª–ª–∞—Ä–æ–≤.")
print(f"–í 2013 –≥–æ–¥—É –±—ã–ª–æ {rounds_2013} —Ä–∞—É–Ω–¥–æ–≤, –∞ —Ç–∏–ø–∏—á–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Å—Ä–µ–¥—Å—Ç–≤ —Å–æ—Å—Ç–∞–≤–∏–ª {typical_amount_2013} –¥–æ–ª–ª–∞—Ä–æ–≤.")
```

* –û—Ü–µ–Ω–∏—Ç–µ, –≤–æ–∑–º–æ–∂–Ω–æ –ª–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–∏—Ç—å –∫ —ç—Ç–∏–º —Ç–∞–±–ª–∏—Ü–∞–º –µ—â—ë –∏ —Ç–∞–±–ª–∏—Ü—É `degrees.csv`.

   - –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–æ **267,694 —Å—Ç—Ä–æ–∫**, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç —É—Å–ø–µ—à–Ω–æ–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞—Ö –∏–∑ **`people.csv`** –∏ –æ–± –∏—Ö –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ –∏–∑ **`education.csv`**.

   - **`network_username`** —Ç–∞–∫–∂–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –º–Ω–æ–≥–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤.

### –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–π –∏ –Ω–µ—Å–æ–≤–ø–∞–¥–µ–Ω–∏–π:

- **–ü–æ –≤—Ç–æ—Ä–æ–º—É –≤–∞—Ä–∏–∞–Ω—Ç—É** (–æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ **`id`** –∏ **`object_id`**):
  - **93,525 —Å–æ–≤–ø–∞–¥–µ–Ω–∏–π**: –°–æ—Ç—Ä—É–¥–Ω–∏–∫–∏, —É –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –¥–∞–Ω–Ω—ã–µ –≤ –æ–±–µ–∏—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö.
  - **133,184 –Ω–µ—Å–æ–≤–ø–∞–¥–µ–Ω–∏–π**: –°–æ—Ç—Ä—É–¥–Ω–∏–∫–∏, —É –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏.

### –ê–Ω–∞–ª–∏–∑:
1. –í **–ø–µ—Ä–≤–æ–º —Å–ª—É—á–∞–µ** (–æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ **`id`** –∏ **`person_id`**), –º—ã –≤–∏–¥–∏–º –±–æ–ª—å—à–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–π, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –ø–æ —ç—Ç–∏–º –∫–ª—é—á–∞–º.
2. –í–æ **–≤—Ç–æ—Ä–æ–º —Å–ª—É—á–∞–µ** (–æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ **`id`** –∏ **`object_id`**), –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—Å–æ–≤–ø–∞–¥–µ–Ω–∏–π –±–æ–ª—å—à–µ, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –æ—à–∏–±–æ—á–Ω–æ–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∏–∑-–∑–∞ –Ω–µ—Å–æ–≤–ø–∞–¥–µ–Ω–∏–π –≤ –¥–∞–Ω–Ω—ã—Ö.




```python
# –®–∞–≥ 2: –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏
company_employee_count = people.groupby('company_id').size()

# –û–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã –≥—Ä—É–ø–ø
bins = [0, 1, 10, 50, 100, 500, 1000, float('inf')]  # –î–æ–±–∞–≤–∏–ª–∏ –≥—Ä—É–ø–ø—É –¥–ª—è —Å—Ç–∞—Ä—Ç–∞–ø–æ–≤ —Å 1 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–º
labels = ['1', '2-10', '11-50', '51-100', '101-500', '501-1000', '1000+']

# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –∫–æ–º–ø–∞–Ω–∏–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤
company_employee_count_grouped = company_employee_count.groupby(
    pd.cut(company_employee_count, bins=bins, labels=labels)
).count()

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print("\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–ø–∞–Ω–∏–π –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤:")
print(company_employee_count_grouped)
```

```python
# –®–∞–≥ 3: –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤ –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏
merged_data['education_missing'] = merged_data['graduated_at'].isnull()  # True, –µ—Å–ª–∏ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏

# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ company_id –∏ –≤—ã—á–∏—Å–ª—è–µ–º –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤ –∫–∞–∂–¥–æ–π –∫–æ–º–ø–∞–Ω–∏–∏
education_completeness_by_company = merged_data.groupby('company_id')['education_missing'].mean()

# –ò—Å–∫–ª—é—á–∞–µ–º –≥—Ä—É–ø–ø—ã, –≥–¥–µ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö
valid_groups = company_employee_count_grouped[company_employee_count_grouped > 0].index

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø–æ –æ—Å—Ç–∞–≤—à–∏–º—Å—è –≥—Ä—É–ø–ø–∞–º
education_completeness_by_group = education_completeness_by_company.groupby(
    pd.cut(company_employee_count, bins=bins, labels=labels)
).mean().loc[valid_groups]  # –£–±–∏—Ä–∞–µ–º –≥—Ä—É–ø–ø—ã, –≥–¥–µ NaN

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print("\n–°—Ä–µ–¥–Ω—è—è –¥–æ–ª—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø–æ –≥—Ä—É–ø–ø–∞–º:")
print(education_completeness_by_group)
```

### –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π —Å–ø–æ—Å–æ–± –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Å degrees ###

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º 'object_id' –≤ —Å—Ç—Ä–æ–∫–æ–≤—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ–±—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å .str
degrees['object_id'] = degrees['object_id'].astype(str)

# –£–±–∏—Ä–∞–µ–º –ø—Ä–µ—Ñ–∏–∫—Å 'p:' –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ 'object_id' –≤ degrees
degrees['object_id'] = degrees['object_id'].str.replace('p:', '', regex=False)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º 'object_id' –≤ —á–∏—Å–ª–æ–≤–æ–π —Ç–∏–ø
degrees['object_id'] = pd.to_numeric(degrees['object_id'], errors='coerce')

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ 'id_x' (–∏–∑ merged_data) –∏ 'object_id' (–∏–∑ degrees.csv)
merged_data_with_degrees = pd.merge(merged_data, degrees, left_on='id_x', right_on='object_id', how='left')

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print("–ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö –ø–æ—Å–ª–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Å degrees.csv:")
display(merged_data_with_degrees.head())
```

```python
# –í—ã–≤–µ–¥–µ–º –ø—Ä–∏–º–µ—Ä—ã —Å—Ç—Ä–æ–∫, –≥–¥–µ company_id_main –∏ company_id_alt –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç
display(company_and_rounds[company_and_rounds['company_id_main'] != company_and_rounds['company_id_alt']].head(10))

# –ü–æ—Å—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —É –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
unique_main = company_and_rounds['company_id_main'].nunique()
unique_alt = company_and_rounds['company_id_alt'].nunique()

print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ company_id_main: {unique_main}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ company_id_alt: {unique_alt}")
```

```python
# –ó–∞–º–µ–Ω—è–µ–º NaN –≤ company_id_main –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏–∑ company_id_alt
company_and_rounds['company_id_main'].fillna(company_and_rounds['company_id_alt'], inplace=True)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ —Ç–µ–ø–µ—Ä—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ company_id_main
print(f"–û—Å—Ç–∞–ª–æ—Å—å –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ company_id_main: {company_and_rounds['company_id_main'].isna().sum()}")

# –£–¥–∞–ª—è–µ–º company_id_alt, —Ç–∞–∫ –∫–∞–∫ –æ–Ω —Ç–µ–ø–µ—Ä—å –¥—É–±–ª–∏—Ä—É–µ—Ç –æ—Å–Ω–æ–≤–Ω–æ–µ –ø–æ–ª–µ
company_and_rounds.drop(columns=['company_id_alt'], inplace=True)
company_and_rounds.rename(columns={'company_id_main': 'company_id'}, inplace=True)

print("–û–±–Ω–æ–≤–ª—ë–Ω–Ω–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö:")
print(company_and_rounds.info())
```

‚úÖ **–¢–µ–ø–µ—Ä—å —É –≤—Å–µ—Ö –∫–æ–º–ø–∞–Ω–∏–π –µ—Å—Ç—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä.**

üîπ –ü–µ—Ä–µ–π—Ç–∏ –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —ç—Ç–∞–ø—É –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö ‚Äì **–∞–Ω–∞–ª–∏–∑—É –¥–∞—Ç (`founded_year`, `closed_at`) –∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—é –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫**.


```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º founded_year –≤ int, –Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏–º NaN —Ç–∞–º, –≥–¥–µ –∏—Ö –Ω–µ—Ç
company_and_rounds['founded_year'] = pd.to_numeric(company_and_rounds['founded_year'], errors='coerce').astype('Int64')

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(company_and_rounds[['founded_year']].info())
print(company_and_rounds['founded_year'].head(10))
```

üîπ –ü–µ—Ä–µ–π—Ç–∏ –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —ç—Ç–∞–ø—É **–∞–Ω–∞–ª–∏–∑—É —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è –∫–æ–º–ø–∞–Ω–∏–π** –∏ —Å–æ–∑–¥–∞–Ω–∏—é –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫.



```python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ company_id –∏ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
company_funding = company_and_rounds.groupby('company_id').agg(
    total_funding=('funding_total', 'sum'),  # –û–±—â–∞—è —Å—É–º–º–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è
    avg_funding_round=('raised_amount', 'mean'),  # –°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Ä–∞—É–Ω–¥–∞
    total_investments=('investment_rounds', 'sum'),  # –û–±—â–µ–µ —á–∏—Å–ª–æ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã—Ö —Ä–∞—É–Ω–¥–æ–≤
    total_funding_rounds=('funding_rounds', 'sum')  # –û–±—â–µ–µ —á–∏—Å–ª–æ —Ä–∞—É–Ω–¥–æ–≤ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è
).reset_index()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
display(company_funding.head(10))
```

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –º–µ—Ç—Ä–∏–∫–∞—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è
missing_values = company_funding.isna().sum()
print("–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –º–µ—Ç—Ä–∏–∫–∞—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è:")
print(missing_values)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è
print("\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è:")
print(company_funding[['total_funding', 'avg_funding_round', 'total_investments', 'total_funding_rounds']].describe())

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ç–æ–ø-10 –∫–æ–º–ø–∞–Ω–∏–π —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏–µ–º
print("\n–¢–æ–ø-10 –∫–æ–º–ø–∞–Ω–∏–π –ø–æ —Å—É–º–º–∞—Ä–Ω–æ–º—É —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—é:")
display(company_funding.sort_values(by='total_funding', ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(12, 6))
sns.boxplot(x=company_funding['total_funding'])
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ–±—â–µ–≥–æ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è (Total Funding)")
plt.xlabel("–°—É–º–º–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è (–≤ $)")
plt.show()
```

```python
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤ –ø–æ –º–µ—Ç–æ–¥—É IQR
Q1 = company_funding['total_funding'].quantile(0.25)  # –ü–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å (25%)
Q3 = company_funding['total_funding'].quantile(0.75)  # –¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å (75%)
IQR = Q3 - Q1  # –ú–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö

# –ì—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"–ì—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤: –Ω–∏–∂–µ {lower_bound:.2f}, –≤—ã—à–µ {upper_bound:.2f}")

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±—Ä–æ—Å–æ–≤
outliers_count = (company_funding['total_funding'] > upper_bound).sum()
print(f"–ß–∏—Å–ª–æ –∫–æ–º–ø–∞–Ω–∏–π, —É –∫–æ—Ç–æ—Ä—ã—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—ã—à–µ –≥—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤: {outliers_count}")
```

```python
# –£–¥–∞–ª—è–µ–º –ø–æ–ª–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã
final_company_data = final_company_data.drop_duplicates()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(f"–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {final_company_data.shape}")
```

### 3.3. –ö—É–ø–ª–µ–Ω—ã –∑–∞–±–µ—Å–ø–ª–∞—Ç–Ω–æ?

* –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –∫–æ–º–ø–∞–Ω–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ –ø—Ä–æ–¥–∞–Ω—ã –∑–∞ –Ω–æ–ª—å –∏–ª–∏ –∑–∞ –æ–¥–∏–Ω –¥–æ–ª–ª–∞—Ä, –∏ –ø—Ä–∏ —ç—Ç–æ–º –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ —É –Ω–∏—Ö –±—ã–ª –Ω–µ–Ω—É–ª–µ–≤–æ–π –æ–±—â–∏–π –æ–±—ä—ë–º —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è.

* –†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏ –≤–µ—Ä—Ö–Ω—é—é –∏ –Ω–∏–∂–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É –≤—ã–±—Ä–æ—Å–æ–≤ –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ `funding_total` –∏ —É–∫–∞–∂–∏—Ç–µ, –∫–∞–∫–∏–º –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è–º –≥—Ä–∞–Ω–∏—Ü—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç. –£—á–∏—Ç—ã–≤–∞—è –æ–∑–≤—É—á–µ–Ω–Ω—ã–µ –≤—ã—à–µ –ø—Ä–µ–¥–ø–æ—Å—ã–ª–∫–∏ –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —Å–æ–æ–±—Ä–∞–∂–µ–Ω–∏—è, —Å–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –ø–ª–∞–Ω–∏—Ä—É–µ–º–æ–π –∑–∞–∫–∞–∑—á–∏–∫–æ–º –∑–∞–º–µ–Ω—ã –Ω–∏–∂–Ω–µ–π –∏ –≤–µ—Ä—Ö–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤ –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏—è, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–µ –ø–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è–º.

```python
# –§–∏–ª—å—Ç—Ä—É–µ–º –∫–æ–º–ø–∞–Ω–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ –ø—Ä–æ–¥–∞–Ω—ã –∑–∞ $0 –∏–ª–∏ $1 –∏ –∏–º–µ–ª–∏ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏–µ
cheap_acquisitions = filtered_companies[
    (filtered_companies['total_funding'] > 0) &  # –ë—ã–ª–æ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏–µ
    (filtered_companies['total_investments'] <= 1)  # –ü—Ä–æ–¥–∞–Ω—ã –∑–∞ $0 –∏–ª–∏ $1
]

# –í—ã–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–∞–∫–∏—Ö –∫–æ–º–ø–∞–Ω–∏–π
print(f"–ß–∏—Å–ª–æ –∫–æ–º–ø–∞–Ω–∏–π, –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∑–∞ $0 –∏–ª–∏ $1, –Ω–æ –ø–æ–ª—É—á–∏–≤—à–∏—Ö —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏–µ: {cheap_acquisitions.shape[0]}")

# –í—ã–≤–æ–¥–∏–º –ø—Ä–∏–º–µ—Ä—ã
display(cheap_acquisitions.head(10))
```

```python
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
Q1 = filtered_companies['total_funding'].quantile(0.25)  # 25-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å
Q3 = filtered_companies['total_funding'].quantile(0.75)  # 75-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å
IQR = Q3 - Q1  # –ú–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã
lower_percentile = (filtered_companies['total_funding'] <= lower_bound).mean() * 100
upper_percentile = (filtered_companies['total_funding'] >= upper_bound).mean() * 100

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print(f"–ì—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤:")
print(f" - –ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞: ${lower_bound:,.2f}")
print(f" - –í–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞: ${upper_bound:,.2f}")
print(f" - –ü—Ä–æ—Ü–µ–Ω—Ç –∫–æ–º–ø–∞–Ω–∏–π –Ω–∏–∂–µ –Ω–∏–∂–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã: {lower_percentile:.2f}%")
print(f" - –ü—Ä–æ—Ü–µ–Ω—Ç –∫–æ–º–ø–∞–Ω–∏–π –≤—ã—à–µ –≤–µ—Ä—Ö–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã: {upper_percentile:.2f}%")
```
