```python
def analyze_datasets(dfs, names):
    """
    Функция для анализа списка датафреймов.

    Parameters:
    dfs (list of pd.DataFrame): Список датафреймов для анализа.
    names (list of str): Список названий файлов (для отображения).
    """

    for df, name in zip(dfs, names):
        print(f"\n🔍 Анализ датасета: {name}")
        print(f"📊 Размерность: {df.shape[0]} строк, {df.shape[1]} столбцов")
        print("-" * 50)

        # Подсчёт пропусков
        missing_percent = df.isna().mean() * 100
        missing_summary = missing_percent[missing_percent > 0].sort_values(ascending=False)
        if not missing_summary.empty:
            print("⚠️ Пропущенные значения (в %):")
            display(missing_summary.to_frame().T)
        else:
            print("✅ Пропусков нет.")

        # Проверка дубликатов
        duplicate_percent = df.duplicated().mean() * 100
        if duplicate_percent > 0:
            print(f"⚠️ Дубликаты: {duplicate_percent:.2f}% записей.")
        else:
            print("✅ Дубликатов нет.")

        # Проверка пустых столбцов
        empty_columns = df.columns[df.isna().all()]
        if empty_columns.any():
            print(f"⚠️ Полностью пустые столбцы: {list(empty_columns)}")

        # Вывод типов данных и примеров
        print("\n📌 Типы данных и примеры значений:")
        sample_data = df.iloc[0:1].T  # Берем первую строку в транспонированном виде
        sample_data.columns = ["Пример данных"]
        sample_data.insert(0, "Тип данных", df.dtypes)
        display(sample_data)

        print("=" * 80)

# Вызываем функцию для всех загруженных датасетов
analyze_datasets(
    dfs=[acquisition, company_and_rounds, degrees, education, fund, investment, people],
    names=["acquisition", "company_and_rounds", "degrees", "education", "fund", "investment", "people"]
)
```

### 2.1. Раунды финансирования по годам

Задание необходимо выполнить без объединения и дополнительной предобработки на основе датасета `company_and_rounds.csv`.

* Составьте сводную таблицу по годам, в которой на основании столбца `raised_amount` для каждого года указан:

  * типичный размер средств, выделяемый в рамках одного раунда;
  * общее количество раундов финансирования за этот год.

* Оставьте в таблице информацию только для тех лет, для которых есть информация о более чем 50 раундах финансирования.

* На основе получившейся таблицы постройте график, который будет отражать динамику типичного размера средств, которые стартапы получали в рамках одного раунда финансирования.

На основе полученных данных ответьте на вопросы:

* В каком году типичный размер собранных в рамках одного раунда средств был максимален?

* Какая тенденция по количеству раундов и выделяемых в рамках каждого раунда средств наблюдалась в 2013 году?

```python
# Шаг 1: Преобразование столбца 'funded_at' в datetime и извлечение года
company_and_rounds['funded_at'] = pd.to_datetime(company_and_rounds['funded_at'], errors='coerce')

# Проверим, есть ли пустые значения после преобразования
print(company_and_rounds['funded_at'].isnull().sum(), "значений не удалось преобразовать в дату.")

# Извлекаем год, если 'funded_at' корректно преобразован в datetime
company_and_rounds['funding_year'] = company_and_rounds['funded_at'].dt.year

# Убираем строки с пропусками в 'raised_amount' и 'funding_year'
company_and_rounds = company_and_rounds.dropna(subset=['raised_amount', 'funding_year'])

# Проверим результат
display(company_and_rounds.head())
```

```python
# Преобразуем столбец 'founded_at' в datetime, заменяя некорректные значения на NaT
company_and_rounds['founded_at'] = pd.to_datetime(company_and_rounds['founded_at'], errors='coerce')

# Проверим тип данных столбца и наличие пустых значений
print(f"Тип данных столбца 'founded_at': {company_and_rounds['founded_at'].dtype}")
print(f"Количество пропущенных значений в 'founded_at': {company_and_rounds['founded_at'].isnull().sum()}")

# Если необходимо, можно извлечь год из 'founded_at' так же, как и из 'funded_at'
company_and_rounds['founded_year'] = company_and_rounds['founded_at'].dt.year

# Проверим результат
display(company_and_rounds.head())
```

```python
display(company_and_rounds.info())
```

```python
# Заполняем пропуски в 'founded_year' значением 'Не указано'
company_and_rounds['founded_year'].fillna('Не указано', inplace=True)

# Проверим результат
display(company_and_rounds.head())
```

```python
display(company_and_rounds.info())
```

```python
# Шаг 2: Составляем сводную таблицу по годам
funding_by_year = company_and_rounds.groupby('funding_year').agg(
    typical_amount=('raised_amount', 'median'),  # типичный размер средств (медиана)
    total_rounds=('raised_amount', 'size')      # общее количество раундов
)

# Оставляем только годы с более чем 50 раундами
funding_by_year = funding_by_year[funding_by_year['total_rounds'] > 50]

# Выводим полученную сводную таблицу для проверки
funding_by_year.head()
```

```python
# Группируем данные по категории и рассчитываем среднее и стандартное отклонение финансирования
category_stats = (
    filtered_companies.groupby("category_code")["total_funding"]
    .agg(["mean", "std", "count"])
    .sort_values(by="mean", ascending=False)
)

# Фильтруем категории, где достаточно данных (например, > 10 стартапов)
category_stats = category_stats[category_stats["count"] > 10]

# Выводим топ-10 категорий по средней цене стартапов
print("🔝 Топ-10 категорий по средней стоимости стартапов:")
display(category_stats.head(10))

# Выводим топ-10 категорий по разбросу (стандартное отклонение)
print("🔝 Топ-10 категорий с наибольшим разбросом стоимости:")
display(category_stats.sort_values(by="std", ascending=False).head(10))
```
