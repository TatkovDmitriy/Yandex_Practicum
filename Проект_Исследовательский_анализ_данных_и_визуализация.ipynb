В работе используются два датасета с информацией о заведениях общественного питания Москвы, полученные из Яндекс Карт и Яндекс Бизнес за лето 2022 года.

## **Датасет 1: `rest_info.csv`**  
Содержит основную информацию о заведениях.  

**Столбцы:**  
- **`name`** — название заведения.  
- **`address`** — адрес заведения.  
- **`district`** — административный район, в котором находится заведение (например, Центральный административный округ).  
- **`category`** — категория заведения (например, «кафе», «пиццерия» или «кофейня»).  
- **`hours`** — информация о днях и часах работы.  
- **`rating`** — рейтинг заведения по оценкам пользователей Яндекс Карт (максимальная оценка — 5.0).  
- **`chain`** — является ли заведение сетевым (0 — нет, 1 — да).  
- **`seats`** — количество посадочных мест.  

## **Датасет 2: `rest_price.csv`**  
Содержит информацию о ценах в заведениях.  

**Столбцы:**  
- **`price`** — ценовая категория заведения (например, «средние», «ниже среднего», «выше среднего»).  
- **`avg_bill`** — строка, содержащая среднюю стоимость заказа в виде диапазона (например, «Средний счёт: 1000–1500 ₽» или «Цена чашки капучино: 130–220 ₽»).  
- **`middle_avg_bill`** — оценка среднего чека (числовое значение):  
  - Если указан диапазон, берётся медиана двух значений.  
  - Если указано одно число, оно переносится в этот столбец.  
  - Если значения нет или оно не начинается с подстроки «Средний счёт», столбец остаётся пустым.  
- **`middle_coffee_cup`** — оценка стоимости чашки капучино (числовое значение):  
  - Если указан диапазон, берётся медиана двух значений.  
  - Если указано одно число, оно переносится в этот столбец.  
  - Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», столбец остаётся пустым.

```python
import pandas as pd

# Функция нормализации текста (приведение к нижнему регистру, удаление лишних пробелов)
def normalize_text(text):
    if pd.isna(text):
        return ""
    return " ".join(text.lower().strip().split())

# Нормализуем названия и адреса заведений
df['name_normalized'] = df['name'].apply(normalize_text)
df['address_normalized'] = df['address'].apply(normalize_text)

# Проверяем явные дубликаты (полное совпадение)
exact_duplicates = df.duplicated(subset=['name_normalized', 'address_normalized'], keep=False)
print(f"Найдено {exact_duplicates.sum()} возможных дубликатов.")

# Выводим примеры возможных дубликатов
df[exact_duplicates].sort_values(by=['name_normalized', 'address_normalized']).head(10)
```

- Для дальнейшей работы создайте столбец `is_24_7` с обозначением того, что заведение работает ежедневно и круглосуточно, то есть 24/7:
  - логическое значение `True` — если заведение работает ежедневно и круглосуточно;
  - логическое значение `False` — в противоположном случае.

```python
import re

# Функция для проверки круглосуточного режима
def check_24_7(hours):
    # Ищем различные варианты записи (с пробелами, через запятую и т. д.)
    pattern = r"(\bкруглосуточно\b|\b24\s*\/\s*7\b|\b7\s*\/\s*24\b)"
    return bool(re.search(pattern, str(hours), re.IGNORECASE))

# Применяем функцию для создания столбца is_24_7
df['is_24_7'] = df['hours'].apply(check_24_7)

# Проверяем распределение значений
print(df['is_24_7'].value_counts())
```

```python


# Boxplot для поиска выбросов
plt.figure(figsize=(14, 6))
sns.boxplot(x=df['seats'])
plt.title("Boxplot количества посадочных мест")
plt.xlabel("Количество посадочных мест")
plt.grid(True)
plt.show()

# Вывод порогов выбросов (IQR-анализ)
Q1 = df['seats'].quantile(0.25)
Q3 = df['seats'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Порог выбросов: {lower_bound:.2f} - {upper_bound:.2f}")
```

### 🔍 Анализ выбросов в количестве посадочных мест

  - **Нижняя граница:** 8.5  
  - **Верхняя граница:** 132.5  

- **Выводы:**  
  1. Заведения с **менее 9 мест** могут быть аномалиями (например, ошибки в данных или концептуальные заведения без посадочных мест).  
  2. Заведения с **более 132 мест** считаются выбросами. Учитывая, что **максимальное значение — 1288**, у нас есть **сильные выбросы**.  



```python
# Подсчёт выбросов
outliers_low = df[df['seats'] < 9].shape[0]
outliers_high = df[df['seats'] > 132.5].shape[0]

print(f"Число выбросов (менее 9 мест): {outliers_low}")
print(f"Число выбросов (более 132 мест): {outliers_high}")
```

### 📌 Выводы:
1. **Нижние выбросы (менее 9 мест)** могут включать:
   - Ошибки в данных (например, заведения без посадочных мест).
   - Маленькие точки продаж (например, кофейные киоски).
   
2. **Верхние выбросы (более 132 мест)** включают:
   - Крупные рестораны и фуд-корты.
   - Ошибки при вводе данных (например, лишние нули).
   
3. **Почти 18% данных — выбросы.** Их удаление может повлиять на анализ, поэтому перед очисткой стоит оценить их влияние.

📌 **Далее:** Заменяем отсутствие данных на среднее значения по категории*.


```python
# Функция замены выбросов на медиану по категории
def replace_outliers(row):
    if row['seats'] < 9 or row['seats'] > 132:
        return df.groupby('category')['seats'].transform('median')[row.name]
    return row['seats']

# Применяем замену
df['seats'] = df.apply(replace_outliers, axis=1)

# Проверяем, остались ли выбросы
outliers_low_after = df[df['seats'] < 9].shape[0]
outliers_high_after = df[df['seats'] > 132].shape[0]

print(f"Число выбросов (менее 9 мест) после замены: {outliers_low_after}")
print(f"Число выбросов (более 132 мест) после замены: {outliers_high_after}")
```

### Задача 6

Изучите, с какими данными показывают самую сильную корреляцию рейтинги заведений? Постройте и визуализируйте матрицу корреляции рейтинга заведения с разными данными: его категория, положение (административный район Москвы), статус сетевого заведения, количество мест, ценовая категория и признак, является ли заведения круглосуточным. Выберите самую сильную связь и проверьте её.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Частотное кодирование для всех категориальных столбцов
categorical_columns = ['category', 'district', 'price']
for col in categorical_columns:
    df[f'{col}_encoded'] = df[col].map(df[col].value_counts(normalize=True))

# Выбираем признаки для корреляции
correlation_df = df[['rating', 'category_encoded', 'district_encoded', 'chain',
                     'seats', 'price_encoded', 'is_24_7']]

# Рассчитываем матрицу корреляции
correlation_matrix = correlation_df.corr()

# Сортируем признаки по корреляции с рейтингом
correlation_with_rating = correlation_matrix['rating'].drop('rating').sort_values(ascending=False)

# Выводим матрицу корреляции
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Матрица корреляции рейтинга с ключевыми параметрами")
plt.xticks(rotation=45)
plt.show()

# Выводим корреляцию рейтинга с другими признаками
print("Корреляция рейтинга с другими признаками:")
print(correlation_with_rating)
```

   - `seats` (количество мест): **0.088** → очень слабая положительная корреляция, возможно, более крупные заведения получают чуть более высокие оценки.

2. **Отрицательная корреляция**:
   - `price_encoded` (ценовая категория): **-0.128** → слабая отрицательная связь, возможно, более дорогие заведения оцениваются ниже.
   - `is_24_7` (круглосуточность): **-0.098** → круглосуточные заведения имеют слегка более низкие рейтинги, что может указывать на проблемы с качеством обслуживания в ночное время.
